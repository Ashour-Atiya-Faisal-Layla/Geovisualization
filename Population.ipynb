```python
# geopandas is not available on Google Colab by default,
#   so we need to bring it in using conda
# NOTE: This may take several minutes! Start it now, and please be patient!
!pip install -q condacolab
import condacolab

condacolab.install()
!mamba install geopandas
!mamba install geoplot

# Do the relevant imports of pandas, geopandas, and geoplot
import pandas as pd
import geopandas as gpd
import geoplot as gplt
import matplotlib.pyplot as plt

# Import shutil to unpack our zipped data
import shutil


!wget https://www.abs.gov.au/census/find-census-data/geopackages/download/Geopackage_2021_G01_AUST_GDA2020.zip
shutil.unpack_archive('Geopackage_2021_G01_AUST_GDA2020.zip')
gda2020 = gpd.read_file('G01_AUST_GDA2020.gpkg')

# This code produces the vis shown at the top of this question
gplt.choropleth(gda2020,
    hue='Tot_P_P',
    cmap='inferno', linewidth=0.5,
    edgecolor='gray',
    legend=True,
    projection=gplt.crs.Mercator()
)
plt.title("Total population, 2020")

```

```python
gda2020.head()
```

```python
F_P = gda2020.Tot_P_F / gda2020.Tot_P_P * 100
gda2020['F_P'] = F_P 
```

```python
M_P = gda2020.Tot_P_M / gda2020.Tot_P_P * 100
gda2020['M_P'] = M_P
```

```python
gda2020.head()
```

```python


gplt.choropleth (
    gda2020,
    hue='F_P',
    cmap='RdPu',
     linewidth=0.5,
    edgecolor='black',
    legend=True, 
)

plt.title("Female ratio in each state , 2020")


gplt.choropleth(
    gda2020,
    hue='M_P',
    cmap= 'GnBu', 
    linewidth=0.5,
    edgecolor='black',
    legend=True,  
)  
plt.title("Male ratio in each state , 2020")
```
